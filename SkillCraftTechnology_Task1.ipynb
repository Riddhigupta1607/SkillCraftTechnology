```python
#filtering the data for a specific indicator and year
indicator_name = 'Population, total' 
year = '2020'
filtered_data = df[df['Indicator Name'] == indicator_name] 
```

```python
print(filtered_data)
```

```python
filtered_data = filtered_data.dropna(subset=[year]) #dropping rows for missing values of a chosen year
```

```python
print(filtered_data)
```

```python
filtered_data = filtered_data.sort_values(by=year, ascending=False).head(10) #Sort data by the year values
```

```python
filtered_data
```

```python
#Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(filtered_data['Country Name'], filtered_data[year].astype(float), color='black')
plt.xlabel('Country')
plt.ylabel(indicator_name)
plt.title(f'{indicator_name} in {year}')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
